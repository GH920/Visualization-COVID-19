```python
import pandas as pd
```

```python
confirmed = pd.read_csv('time_series_covid_19_confirmed.csv')
confirmed.head()
```

```python
deaths = pd.read_csv('time_series_covid_19_deaths.csv')
deaths.head()
```

```python
recovered = pd.read_csv('time_series_covid_19_recovered.csv')
recovered.head()
```

```python
confirmed_unpivot = confirmed.groupby('Country/Region').sum().reset_index().melt(id_vars=confirmed.columns[1:4], var_name='Date', value_name='Confirmed')
deaths_unpivot = deaths.groupby('Country/Region').sum().reset_index().melt(id_vars=deaths.columns[1:4], var_name='Date', value_name='Deaths')
recovered_unpivot = recovered.groupby('Country/Region').sum().reset_index().melt(id_vars=recovered.columns[1:4], var_name='Date', value_name='Recovered')
```

```python
join_on = ['Country/Region','Date']
confirmed_deaths = pd.merge(confirmed_unpivot, deaths_unpivot.drop(['Lat','Long'],axis=1), left_on=join_on, right_on=join_on)
```

```python
confirmed_deaths_recovered = pd.merge(confirmed_deaths, recovered_unpivot.drop(['Lat','Long'],axis=1), left_on=join_on, right_on=join_on)
```

```python
confirmed_deaths_recovered.head()
```

```python
confirmed_deaths_recovered.to_csv('confirmed_deaths_recovered.csv')
```

```python
def integrate(confirmed):
    confirmed.loc[confirmed['Country/Region']=='Taiwan*','Province/State'] = 'Taiwan'
    confirmed.loc[confirmed['Country/Region']=='Taiwan*','Country/Region'] = 'China'
```

```python
integrate(confirmed)
integrate(deaths)
integrate(recovered)
```

```python
confirmed_China = confirmed[confirmed['Country/Region'] == 'China']
```

```python
deaths_China = deaths[deaths['Country/Region'] == 'China']
```

```python
recovered_China = recovered[recovered['Country/Region'] == 'China']
```

```python
confirmed_China.to_csv('confirmed_China.csv')
deaths_China.to_csv('deaths_China.csv')
recovered_China.to_csv('recovered_China.csv')
```

```python
def calc_growth(confirmed_China):
    timeline = confirmed_China.columns[4:]
    growth_confirmed_China = confirmed_China.iloc[:,:-1]
    for i in range(len(timeline)-1):
        growth_confirmed_China[timeline[i]] = confirmed_China[timeline[i+1]] - confirmed_China[timeline[i]]
    return growth_confirmed_China
```

```python
growth_confirmed = calc_growth(confirmed)
growth_deaths = calc_growth(deaths)
growth_recovered = calc_growth(recovered)
```

```python
growth_confirmed.head()
```

```python
growth_confirmed_unpivot = growth_confirmed.groupby('Country/Region').sum().reset_index().melt(id_vars=growth_confirmed.columns[1:4], var_name='Date', value_name='Confirmed')
growth_deaths_unpivot = growth_deaths.groupby('Country/Region').sum().reset_index().melt(id_vars=growth_deaths.columns[1:4], var_name='Date', value_name='Deaths')
growth_recovered_unpivot = growth_recovered.groupby('Country/Region').sum().reset_index().melt(id_vars=growth_recovered.columns[1:4], var_name='Date', value_name='Recovered')
```

```python
growth_confirmed_deaths = pd.merge(growth_confirmed_unpivot, growth_deaths_unpivot.drop(['Lat','Long'],axis=1), left_on=join_on, right_on=join_on)
growth_confirmed_deaths_recovered = pd.merge(growth_confirmed_deaths, growth_recovered_unpivot.drop(['Lat','Long'],axis=1), left_on=join_on, right_on=join_on)
```

```python
growth_confirmed_deaths_recovered.to_csv('growth_confirmed_deaths_recovered.csv')
```
